# 1: Introduction

## 1.1 Problem Statement

Developing a competitive NIH research proposal is a time‑consuming and cognitively demanding process. As described in the project brief, Dr. Ian McCulloh faces the challenge of aligning his extensive and diverse research background with a specific NOFO focused on digital mental health solutions. The volume of his prior work makes it difficult to manually identify the most relevant findings, themes, and methodological strengths. As the brief states:

> “The sheer volume of his past work makes it challenging to manually sift through for relevant information that could aid in developing a compelling grant proposal.”

This project addresses that challenge by building an AI‑powered system that automates the extraction of relevant insights, aligns them with NOFO requirements, and generates structured NIH‑style proposal components.

## 1.2 Project Goals

The goal of this project is to create a generative AI system that can support end‑to‑end NIH proposal development. Specifically, the system aims to:

+ Automate NIH proposal drafting using modular LLM pipelines
+ Identify relevant prior research using topic extraction and relevance scoring
+ Generate new project ideas aligned with the NOFO
+ Produce Specific Aims and a structured proposal blueprint
+ Evaluate proposal alignment with NOFO criteria
+ Provide polishing and reviewer‑style feedback
+ Support PDF ingestion, FAISS retrieval, and a Streamlit UI

These goals directly reflect the project description:

> “Automate and optimize the proposal drafting process… identify relevant past research findings… generate high‑quality proposal content satisfying both technical and scientific standards.”

1.3 System Overview

The system follows the workflow outlined in the project brief: read the NOFO, analyze research papers, filter by topic, generate ideas, and iteratively refine proposal content. The architecture includes:

+ Ingestion layer (PDF/text loading + chunking)
+ Retrieval layer (embeddings + FAISS index + caching)
+ LLM pipeline (topics → relevance → gaps → ideas → aims → blueprint)
+ Formatting layer (NIH‑style narrative)
+ Evaluation layer (NOFO alignment + reviewer simulation)
+ UI layer (Streamlit single‑page + multipage app)

This modular design ensures clarity, extensibility, and alignment with the rubric’s step‑based evaluation.

# 15: Summary and Recommendations

Summarize:

This project implements an end‑to‑end NIH proposal assistant that closely mirrors the workflow described in the rubric: topic extraction, research paper relevance assessment, proposal ideation, proposal blueprint preparation, evaluation against NOFO criteria, and final summary and recommendations. Starting from a corpus of prior research, the system ingests text and PDFs, uses FAISS‑based retrieval to focus on relevant content, and then chains LLM modules to generate topics, assess alignment with a target direction, identify gaps, propose new ideas, draft Specific Aims, and assemble a coherent blueprint. The final outputs include a formatted NIH‑style proposal, a polished narrative, and simulated reviewer feedback.

In practice, the pipeline performs well at surfacing plausible topics and generating structured proposal components, especially when the input corpus is reasonably aligned with the target NOFO. The FAISS retrieval and caching help keep the system responsive, and the Streamlit UI makes it usable beyond the notebook. However, the system still depends heavily on prompt quality, corpus relevance, and careful human review.

Recommendations for improvement include:

+ Tightening prompts to reduce redundancy and hallucination
+ Adding more explicit NOFO‑aware constraints in the blueprint and polisher stages
+ Improving PDF extraction quality for complex documents
+ Integrating more robust evaluation metrics (e.g., automatic checks for NIH section coverage, word limits, and alignment with specific review criteria).

It would also be valuable to log intermediate outputs systematically so that users can trace how each proposal component was derived from the underlying corpus.

# 16 Reflections

## 16.1 Limitations

+ Model hallucination: Even with retrieval grounding, the LLM may introduce unsupported details.
+ FAISS retrieval sensitivity: Retrieval quality depends heavily on chunking, embedding model, and k‑value.
+ PDF extraction noise: Multi‑column or figure‑heavy PDFs reduce extraction quality.
+ NIH formatting complexity: The system generates NIH‑style sections but does not enforce all NIH constraints (page limits, biosketch integration, etc.).

## 16.2 Future Work

+ Multi‑agent reviewer panel to simulate diverse study section perspectives
+ Structured scoring rubric aligned with NIH criteria
+ Integration with NIH ASSIST for export‑ready proposal sections
+ Fine‑tuned domain models for mental health and NIH proposal language
+ Automated NOFO compliance checker

## 16.3 Lessons Learned
This project highlighted the importance of modular pipeline design, careful prompt engineering, and high‑quality data ingestion. It also reinforced that AI systems for proposal automation are most effective when they augment expert judgment rather than replace it. The iterative workflow—topic extraction → relevance → gaps → ideas → aims → blueprint—mirrors real proposal development and demonstrates how generative AI can accelerate early‑stage ideation and structuring.
